## Normal Distribution

1. In the same way in the Uniform Distribution challenge, write a function that generates normally distributed random numbers.
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 1
1. Generate 1,000 normally distributed numbers with the average of 10 and standard deviation of 50
2. Plot the distributions of the data generated.

Expected output:

![normal distribution](nd.png)

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_normal_distribution(mean, std_dev, size):
    # Generate normally distributed numbers
    return np.random.normal(mean, std_dev, size)

# Generate 1,000 normally distributed numbers with mean 10 and standard deviation 1
data_std_1 = generate_normal_distribution(mean=10, std_dev=1, size=1000)

# Generate 1,000 normally distributed numbers with mean 10 and standard deviation 50
data_std_50 = generate_normal_distribution(mean=10, std_dev=50, size=1000)

# Plot the distributions
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data_std_1, bins=30, color='blue', edgecolor='black')
plt.title('Normal Distribution (=10, =1)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(data_std_50, bins=30, color='blue', edgecolor='black')
plt.title('Normal Distribution (=10, =50)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

## Exponential Distribution

1. Using `numpy.random.exponential`, create a function that returns a list of numbers exponentially distributed with the mean of 10. 

1. Use the function to generate two number sequences with the size of 10 and 100.

1. Plot the distributions as histograms with the nubmer of bins as 100.

Your output should look like below:

![exponential distribution](ed.png)

```python
## Exponential Distribution of Real Data

import numpy as np
import matplotlib.pyplot as plt

def generate_exponential_distribution(size, mean):
    # The scale parameter (beta) is equal to the mean for an exponential distribution
    scale = mean
    # Generate exponentially distributed numbers
    return np.random.exponential(scale, size)

# Generate two sequences with sizes 10 and 100
sequence_10 = generate_exponential_distribution(10, 10)
sequence_100 = generate_exponential_distribution(100, 10)

# Plot the distributions as histograms
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(sequence_10, bins=100, color='blue', edgecolor='black')
plt.title('Exponential Distribution (Size 10)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(sequence_100, bins=100, color='blue', edgecolor='black')
plt.title('Exponential Distribution (Size 100)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

How are the two distributions different?

The mean changes, so the distribution changes as well. 

The two histograms represent exponential distributions of different sample sizes (10 and 100). Here are the key differences between the two distributions:

1. Sample Size:
   - The left histogram represents a sample size of 10.
   - The right histogram represents a sample size of 100.

2. Shape and Smoothness:
   - The histogram with a sample size of 10 appears more irregular and less smooth. This is typical for smaller sample sizes, where the randomness and variability in the data are more pronounced.
   - The histogram with a sample size of 100 is smoother and more continuous. Larger sample sizes tend to provide a better approximation of the underlying distribution, reducing the impact of random fluctuations.

3. Frequency Distribution:
   - In the histogram with a sample size of 10, the frequency for each bin is either 0 or 1, as there are only 10 data points. This results in a more "spiky" appearance.
   - In the histogram with a sample size of 100, the frequencies are higher and more varied, leading to a more detailed and nuanced representation of the distribution.

4. Representation of the Exponential Distribution**:
   - Both histograms aim to represent an exponential distribution with a mean of 10. However, the larger sample size (100) provides a clearer and more accurate depiction of the exponential decay characteristic of this distribution.
   - The histogram with a sample size of 100 shows a higher concentration of values near zero, tapering off as the value increases, which is typical for an exponential distribution.

In summary, the primary difference lies in the sample size, which affects the smoothness, shape, and representation of the exponential distribution. Larger sample sizes generally provide a more accurate and smooth representation of the underlying distribution.


## Exponential Distribution of Real Data

Suppose that the amount of time one spends in a bank is exponentially distributed with mean as 10 minutes (i.e.  = 1/10). What is the probability that a customer will spend less than fifteen minutes in the bank? 

Write a code in python to solve this problem

```python
import scipy.stats as stats

# Parameters
lambda_ = 1/10  # As the mean is 10, lambda = 1/mean
time_less_than = 15

# Calculate the probability using the exponential CDF
probability = stats.expon.cdf(time_less_than, scale=1/lambda_)

# Round to three decimal places for clarity
probability_rounded = round(probability, 3)

print(f"Probability that a customer will spend less than {time_less_than} minutes in the bank: {probability_rounded}")
```
